###  Paired Sample t-test

We want to Analyze whether an intervention such as an event/promition has changed the population parameter


```python
# https://www.dropbox.com/s/8nhnpipewos7u88/breakups.csv?dl=0
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!wget https://www.dropbox.com/s/8nhnpipewos7u88/breakups.csv
```

```python
breakups_df = pd.read_csv("breakups.csv")
breakups_df.head()
```

```python
sns.distplot(breakups_df["Before_Breakup"],label = "Before Breakup")
sns.distplot(breakups_df["After_Breakup"],label = "After Breakup")
plt.legend()
plt.show()
```

```python
from scipy import stats
```

```python
stats.ttest_rel(breakups_df["Before_Breakup"], breakups_df["After_Breakup"])
```

```python
# null hypothesis follows - no impact of breakup on Alchohol consumption
```

# Anova - Analysis of Variance

## About Anova
We use this ANOVA testing when we want to conduct an hypothesis test to compare mean values simultaneously for more than two groups and these two groups are created from the same samples using a factor

An example can be something as : Let's say I want to analyse a marketing activity and analyse about the sales of the product with  three different discount values so in such scenarios we make use of this ANOVA 



**Problem Statement:**
The brand manager let's say a "surf excel" - A Detergent powder company and they are going to offer the discounts.
-  discount of 0% 
- 10% and 
- 20%  on a randomly selected days.

Now we have the data that contains the quantity sold in kilograms of the detergent powder on each day. As a part of our hypothesis testing we want to perfom one way ANOVA to check whether the discount has any impact on the quantity of detergents sold.

image.png

```python
!wget "https://www.dropbox.com/s/h7zl0jzmrqd1a6e/surf.csv"
```

```python
import pandas as pd
surf_df = pd.read_csv("surf.csv")
surf_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(surf_df["discount_0"],label = "0%")
sns.distplot(surf_df["discount_10"],label = "10%")
sns.distplot(surf_df["discount_20"],label = "20%")
plt.xlabel("Kilogram sold")
plt.legend()
plt.show()
```

```python
from scipy import stats
```

```python
"""
The one-way ANOVA tests the null hypothesis that two or more groups have
the same population mean.  The test is applied to samples from two or
more groups, possibly with differing sizes.
"""
```

```python
stats.f_oneway(surf_df["discount_0"],
               surf_df["discount_10"],
               surf_df["discount_20"])
```
