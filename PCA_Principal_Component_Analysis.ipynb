```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
digit = load_digits()
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(digit.data)
df.head()
```

```python
df.shape
```

```python
df["target"] = digit.target
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
digits_new = ss.fit_transform(digit.data)
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
digit_pca = pca.fit_transform(digits_new)
digit_pca
```

```python
digit_pca.shape
```

```python
np.round(pca.explained_variance_ratio_,2)
```

```python
pca = PCA(n_components=18)
digit_pca = pca.fit_transform(digits_new)
np.round(pca.explained_variance_ratio_,2)
```

```python
np.sum(np.round(pca.explained_variance_ratio_,2))
```

```python
plt.figure(figsize=(20,5))
plt.plot(range(digit_pca.shape[1]), np.cumsum(pca.explained_variance_ratio_))
plt.xticks(range(digit_pca.shape[1]))
plt.title("explained variance")
plt.grid(True)
plt.show()
```
